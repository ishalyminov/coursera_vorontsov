```python
import pandas as pd

from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
TRAINSET_FILE = '_3abd237d917280ba0d83bfe6bd49776f_perceptron-train.csv'
TESTSET_FILE = '_3abd237d917280ba0d83bfe6bd49776f_perceptron-test.csv'

trainset = pd.read_csv(TRAINSET_FILE, header=None)
testset = pd.read_csv(TESTSET_FILE, header=None)
trainset
```

```python
X_train = trainset.drop(0, 1)
X_test = testset.drop(0, 1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

y_train = trainset[0]
y_test = testset[0]
```

```python
# unscaled
clf = Perceptron(random_state=241)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, predictions)
accuracy_unscaled
```

```python
# scaled
clf = Perceptron(random_state=241)
clf.fit(X_train_scaled, y_train)
predictions = clf.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, predictions)
accuracy_scaled
```

```python
with open('task_1', 'w') as task_1_out:
    task_1_out.write('{0:.3f}'.format(accuracy_scaled - accuracy_unscaled))
```
